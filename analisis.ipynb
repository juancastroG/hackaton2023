```python
import pandas as pd
import numpy as np

dataset1 = pd.read_csv('Emergencias_2018_-_UNGRD.csv')
dataset1.head()
```

```python
dataset1['TIPO DE EVENTO'].unique()
```

```python
dataset1=dataset1[['FECHA','TIPO DE EVENTO','FALLECIDOS','HERIDOS','AFECTADOS','FAMILIAS','DEPARTAMENTO','MUNICIPIO']]
dataset1 = dataset1[dataset1['TIPO DE EVENTO']=='INUNDACIÓN']
dataset1.head()
```

```python
#cantidad de datos por departamento
dataset1['DEPARTAMENTO'].value_counts()
```

```python
#filtarr solo departamento de antioquia
dataset1 = dataset1[dataset1['DEPARTAMENTO']=='BOYACA']
dataset1.head()
```

```python
#municio de antioquia
dataset1['MUNICIPIO'].unique()
```

```python
#analisis rapido seewtviz 
import sweetviz as sv
report = sv.analyze(dataset1)
report.show_notebook()

```

```python
#de la fecha extrae el año, el formato es el siguiente = '01/03/2018 12:00:00 AM	'
dataset1['año'] = dataset1['FECHA'].str[6:10] 
dataset1['mes'] = dataset1['FECHA'].str[0:2] 
dataset1.head()
#contar cuantos eventos hay por año

#organizar por mes
dataset1 = dataset1.sort_values(by=['mes'])
dataset1.head()
dataset1['mes'].value_counts()
```

```python
#Grafica de casos por mes con plotly
import plotly.express as px
fig = px.bar(dataset1, x='mes', y='FAMILIAS',color='año',title='Casos de inundación por mes')
fig.show()

```

```python
# ahora el numero de afectados por mes y mostrarlos en un dTAFRAME
dataset1['AFECTADOS'].groupby(dataset1['mes']).sum()
```
